```python
from PIL import Image
from matplotlib import image
from matplotlib import pyplot
import matplotlib.pyplot as plt
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#image1=Image.open('C:/Users/ESTUDIANTE/Downloads/MULTI/peppers.png')
#image.show()

image2=cv2.imread('PELOTASLISAS-COLORES.jpg')
rgb2=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)

#gray=cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)
#YCrCb=cv2.cvtColor(image2,  cv2.COLOR_BGR2HLS)
hsv=cv2.cvtColor(image2,  cv2.COLOR_BGR2HSV)

plt.axis('off')
plt.imshow(rgb2)
plt.show()
```

# **Segmentación de cada una de las 6 pelotas, de manera independiente.**

```python
#Blanco
lower_B = np.array([0,5,170])
upper_B = np.array([179,20,255])
mask_B=cv2.inRange(hsv,lower_B,upper_B)
RGB_B=cv2.cvtColor(mask_B,cv2.COLOR_BGR2RGB)


plt.axis('off')
plt.imshow(RGB_B)
plt.show()

white=cv2.bitwise_and(image2,image2,mask=mask_B)
RGB2_B=cv2.cvtColor(white,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(RGB2_B)
plt.show()
```
